## This file train a LinearRegression model with poly=2. Then by predicted values of Lambda, predict the path of fixation for Moran process in Erdosh-Reyni graph. The final results are path of fixation for ER graph with size 3000, 5000, and 10000, which are chagable for larger graph. The file ER.csv was build in file 'Generating ER'.

```python
import numpy as np 
import random
import pandas as pd
import networkx as nx
import collections
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
def lamda(n,i,r):
    return(i*r*(n-i)/((n-i+i*r)*(n-1)))
```

```python
df=pd.read_csv('/home/hajihashemi/Desktop/AAA/ER.csv')  
df = df.drop(['Unnamed: 0'], axis=1)
df.columns
```

```python
rel=list(df['rel'])
fit=list(df['fit'])
size=list(df['size'])
tar=list(df['target'])
pro=list(df['pro'])

from sklearn.tree import DecisionTreeRegressor

# Assuming you have your data in the following format:
# X: array-like or DataFrame with shape (n_samples, n_features)
# Y: array-like with shape (n_samples,)

# Convert the lists into a numpy array
X = np.array([rel,fit,pro]).T
Y = np.array(tar)

from sklearn.preprocessing import PolynomialFeatures
cern_poly = PolynomialFeatures(2)
X = cern_poly.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.05, random_state=42)

# Create a linear regression model
model = LinearRegression()
#model=RandomForestRegressor(  )

# Fit the model to the training data
model.fit(X_train, Y_train)

# Predict TAR values for the training and test sets
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

# Calculate MSE for the training set
train_mse = mean_squared_error(Y_train, train_predictions)

# Calculate MSE for the test set
test_mse = mean_squared_error(Y_test, test_predictions)

# Print the MSE values
print("MSE for the training set:", train_mse)
print("MSE for the test set:", test_mse)
```

```python
#[rel, fit,em,pro,size]
def lamda2(n,i,r):
    Q=(float(model.predict(cern_poly.fit_transform([[i/n, r]]))))*(lamda(n,i,r))
    return(Q)
    
```

```python
def N(n,r,i,j):
    A=(r**j-1)
    B=((r**(n-i))-1)
    C=(r**n-1)
    D=(r**i-1)
    F=(r**n-1)
    G=r**(i-j)
    K=((r**(n-j))-1)
    if j<=i:
        R=A**2*B*G/((r-1)*C*D)
    else:
        R=K*A/((r-1)*F)
    return(R)

def NF(n,r,i,j):
    return(N(n,r,i,j)*r/lamda2(n,j,r))

def AN(n,k,r):
    t=0
    for j in range (1,k):
        t+=NF(n,3,1,j)-NF(n,3,k,j)
    return(t)
```
